# BBC FY22 Q4

```python
bbc1=r'/kaggle/input/bbc-q4/BBCFY21Q4.csv'
bbc2=r'/kaggle/input/bbc-q4/BBCFY22Q4.csv'
bbcActual=r'/kaggle/input/bbc-q4/BBCFY22Q4Actual.csv'
```

```python
import pandas as pd

bbc1=pd.read_csv(bbc1)
bbc1.head()
```

```python
from babel.numbers import format_currency

bbc1=bbc1[['GL Account', 'GL Description', 'Allocated Budget', 'Actual', '% Used']]
bbc1.rename(columns={'Allocated Budget': 'Budget', 'GL Description': 'Revenue'}, inplace=True)
bbc1.insert(5, "", " ") 
bbc1.head()
```

# BBC FY21 Q4

```python
#READ PREVIOUS QRT DATA:
bbc2=pd.read_csv(bbc2)
bbc2=bbc2[['GL Account', 'Allocated Budget', 'Actual', '% Used']]
bbc2.head()
```

```python
bbc2.rename(columns={'Allocated Budget': ' Budget', 'Actual': ' Actual','% Used': ' % Used'}, inplace=True)
bbc2.head()
```

# Actuals

```python
bbcActual=pd.read_csv(bbcActual)
bbcActual.head()
```

```python
bbcActual=bbcActual[['GL Account', ' Closing Balance']]
bbcActual=bbcActual.rename(columns={' Closing Balance':'Actuals Through June'})
bbcActual.head()
```

# Join Data with Previous Year Data

```python
bbc=bbc1.merge(bbc2, on='GL Account').merge(bbcActual, on='GL Account')
bbc.head()
```

```python
bbc=bbc.rename(columns={'GL Account':'Object Code'})
bbc.head()
```

# Split Revenue and Expense

```python
#bbc['dept']  =   bbc['Object Code'].astype(str).str[9:14]
bbc['Object Code']  = bbc['Object Code'].astype(str).str[-6:].astype(int)
bbc.head(7)
```

#### Revenue

```python
import warnings
warnings.filterwarnings("ignore")

intCols=['Budget','Actual', ' Budget', ' Actual', 'Actuals Through June']
```

```python
revDF=bbc[bbc['Object Code'] < 600_000].sort_values(by='Object Code')
revDF.loc['Total']= revDF[intCols].sum()

revDF["Budget"] =  revDF["Budget"]*-1
revDF[" Budget"] =  revDF[" Budget"]*-1
revDF["Actual"] = revDF["Actual"]*-1
revDF[" Actual"] = revDF[" Actual"]*-1
revDF["Actuals Through June"] = revDF["Actuals Through June"]*-1

revDF["Budget"] =  revDF["Budget"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
revDF[" Budget"] = revDF[" Budget"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
revDF["Actual"] = revDF["Actual"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
revDF[" Actual"] = revDF[" Actual"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
revDF["Actuals Through June"] = revDF["Actuals Through June"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
revDF.head()
```

#### Expenses

```python
expDF=bbc[bbc['Object Code']>600_000].sort_values(by='Object Code')
expDF = expDF.rename(columns={'Revenue': 'Expenses'})
expDF.loc['Total']= expDF[intCols].sum()

expDF["Budget"] = expDF["Budget"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
expDF[" Budget"] = expDF[" Budget"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
expDF["Actual"] = expDF["Actual"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
expDF[" Actual"] = expDF[" Actual"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
expDF["Actuals Through June"] = expDF["Actuals Through June"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
expDF.head()
```
