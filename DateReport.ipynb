# Date Report

#### Enter the Dates to Generate Report

```python
date1='7/01/2021'
date2='9/30/2021'
dept='Food'
```

#### Change Nothing Below:

```python
import numpy as np 
import pandas as pd 
import os
from datetime import datetime
import time

start=time.time()
date=datetime.today().strftime('%m-%d-%Y')
print("Running Reggie Report on: ", date)

file=r'/kaggle/input/datareport/dataReport.csv'
df=pd.read_csv(file)
df.head()
```

```python
df=df[['GL Account', ' Date', ' YTD Debits', ' YTD Credits']]
df.head()
```

```python
df[[' YTD Debits', ' YTD Credits']]=df[[' YTD Debits', ' YTD Credits']].fillna(0)
df.head()
```

```python
df['DataTotal'] = df.apply((lambda x: x[' YTD Debits'] - x[' YTD Credits']), axis=1)
df.head()
```

```python
df=df.drop(columns=[' YTD Debits', ' YTD Credits'])
df.head()
```

```python
df=df.groupby(['GL Account', ' Date'])['DataTotal'].sum().reset_index()
df.head()
```

```python
df=df[(df[' Date'] > date1) & (df[' Date'] < date2)].reset_index()
df.head()
```

```python
df=df.groupby(['GL Account'])['DataTotal'].sum().reset_index()
df.head(1000)
```

```python
df1=df[(df['GL Account'] > 200) & (df['GL Account'] < 300)]
df1.head()
```

```python
df2=df[(df['GL Account'] > 300) & (df['GL Account'] < 400)]
df2.head()
```

```python
df3=df[(df['GL Account'] > 400) & (df['GL Account'] < 500)]
df3=df3.style.set_caption("Hello World")
df3
```
