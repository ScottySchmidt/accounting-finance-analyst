# Qrt Report on one Tab

```python
import pandas as pd
admin1=r'/kaggle/input/finalqrt/GLBA.csv'
admin1=pd.read_csv(admin1)
admin1.head()
```

```python
from babel.numbers import format_currency

admin1=admin1[['GL Account', 'GL Description', 'Allocated Budget', 'Actual', '% Used']]
admin1.rename(columns={'Allocated Budget': 'Budget', 'GL Description': 'Revenue'}, inplace=True)
admin1.insert(5, " ", " ") 
admin1.head()
```

# Previous Year Data

```python
#READ PREVIOUS QRT DATA:
admin2=r'/kaggle/input/finalqrt/GLBA1.csv'
admin2=pd.read_csv(admin2)
admin2=admin2[['GL Account', 'Allocated Budget', 'Actual', '% Used']]
admin2.head()
```

```python
admin2.rename(columns={'Allocated Budget': 'Budget1', 'Actual': 'Actual1','% Used': '% Used1'}, inplace=True)
admin2.head()
```

# Join Data with Previous Year Data

```python
data=admin1.merge(admin2, on='GL Account')
data.head()
```

```python
data=data.rename(columns={'GL Account':'Object Code'})
data.head()
```

# Split Revenue and Expense

```python
data['dept']  =   data['Object Code'].astype(str).str[9:14]
data['Object Code']  = data['Object Code'].astype(str).str[-6:].astype(int)
data
```

#### Revenue

```python
import warnings
warnings.filterwarnings("ignore")

intCols=['Budget','Actual', '% Used', 'Budget1', 'Actual1', '% Used1']
```

```python
revDF=data[data['Object Code'] < 600_000].sort_values(by='Object Code')
revDF.loc['Total']= revDF[intCols].sum()

revDF["Budget"] =  revDF["Budget"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
revDF["Budget1"] = revDF["Budget1"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
revDF["Actual"] = revDF["Actual"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
revDF["Actual1"] = revDF["Actual1"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
revDF.head()
```

#### Expenses 

```python
expDF=data[data['Object Code']>600_000].sort_values(by='Object Code')
expDF = expDF.rename(columns={'Revenue': 'Expenses'})
expDF.loc['Total']= expDF[intCols].sum()

#expDF["Budget"] = expDF1["Budget"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
#expDF["Budget1"] = expDF["Budget1"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
#expDF["Actual"] = expDF["Actual"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
#expDF["Actual1"] = expDF["Actual1"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
expDF.head()
```
