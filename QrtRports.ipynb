# Quarterly Report GLBA

```python
import pandas as pd
data=r'/kaggle/input/glbaqrt/GLBA.csv'
df=pd.read_csv(data)
df.head()
```

```python
df=df[['GL Account', 'Allocated Budget', 'Actual', '% Used']]
intCols=['Allocated Budget','Actual', '% Used']
df.head()
```

```python
df1=df[(df['GL Account'] > 2_470_000) & (df['GL Account']<2_480_000)]
df1.loc['Total Income']= df1[intCols].sum()
df1.head()
```

```python
df2=df[(df['GL Account'] > 2_450_000) & (df['GL Account'] < 2_490_000)]
df2.loc['Total Income']= df2[intCols].sum()
df2.head()
```

```python
df3=df[(df['GL Account'] > 2_480_000) & (df['GL Account'] < 2_490_000)]
df3.loc['Total Income']= df[intCols].sum()
df3
```
