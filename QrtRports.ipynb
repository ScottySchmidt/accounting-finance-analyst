# Quarterly Report GLBA

# Current Quarter Data

```python
import pandas as pd
data1=r'/kaggle/input/glbaqrt/GLBA.csv'
data1=pd.read_csv(data1)
data1.head()
```

```python
from babel.numbers import format_currency

data1=data1[['GL Account', 'GL Description', 'Allocated Budget', 'Actual', '% Used']].fillna(0)
data1.rename(columns={'Allocated Budget': 'Budget', 'GL Description': 'Revenue'}, inplace=True)
data1.insert(5, " . ", " ") 

#CURRENCY:
#data["Budget"] = data["Budget"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
#data["Actual"] = data["Actual"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
#data['% Used'] = data['% Used'].apply(lambda x: "{0:.2f}%".format(x*100))
data1.head()
```

# Previous Year Data

```python
#READ PREVIOUS QRT DATA:
data2=r'/kaggle//input/glbaq1/GLBA1.csv'
data2=pd.read_csv(data2)
data2=data2[['GL Account', 'Allocated Budget', 'Actual', '% Used']].fillna(0)
data2.head()
```

#### Rename Columns

```python
data2.rename(columns={'Allocated Budget': 'Budget1', 'Actual': 'Actual1','% Used': '% Used1'}, inplace=True)

#CHANGE CURRENCY:
#data2["Budget1"] = data2["Budget1"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
#data2["Actual1"] = data2["Actual1"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
#data2['% Used1'] = data2['% Used1'].apply(lambda x: "{0:.2f}%".format(x*100))
data2.head()
```

# Join Data with Previous Year Data

```python
data=data1.merge(data2, on='GL Account')
data.head()
```

```python
data=data.rename(columns={'GL Account':'Object Code'}).reset_index(drop=True)
data.head()
```

# Split Revenue and Expense

```python
data['dept']  =   data['Object Code'].astype(str).str[:-3].astype(int)
data['Object Code']  = data['Object Code'].astype(str).str[-3:].astype(int)
data.head()
```

```python
revDF=data[data['Object Code'] < 600].sort_values(by='Object Code')
revDF.head()
```

```python
expDF=data[data['Object Code']>600].sort_values(by='Object Code')
expDF = expDF.rename(columns={'Revenue': 'Expenses'})
expDF.head()
```

# Create New Tabs by Dept

```python
import warnings
warnings.filterwarnings("ignore")

intCols=['Budget','Actual', '% Used', 'Budget1', 'Actual1', '% Used1']
revDF1=revDF[(revDF['dept'] ==2471)]
revDF1.loc['Total Income']= revDF1[intCols].sum()

revDF1["Budget"] = revDF1["Budget"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
revDF1["Budget1"] = revDF1["Budget1"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
revDF1["Actual"] = revDF1["Actual"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
revDF1["Actual1"] = revDF1["Actual1"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
#df1['% Used1'] = df1['% Used1'].apply(lambda x: "{0:.2f}%".format(x*100))
revDF1.head()
```

```python
expDF1=expDF[(expDF['dept'] ==2471)]
expDF1.loc['Total Income']= expDF1[intCols].sum()
expDF1.head()
```

```python
df2=revDF[(revDF['dept']==2472)].reset_index(drop=True)
df2.loc['Total Income']= df2[intCols].sum()
df2.head()
```

```python
df3=revDF[(revDF['dept'] == 2473)]
df3.loc['Total Income']= revDF[intCols].sum()
df3.head()
```
