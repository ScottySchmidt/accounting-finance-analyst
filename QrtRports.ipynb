# Quarterly Report GLBA

# Current Quarter Data

```python
import pandas as pd
data=r'/kaggle/input/glbaqrt/GLBA.csv'
joinDF=pd.read_csv(data)
joinDF.head()
```

```python
joinDF=joinDF[['GL Account', 'Allocated Budget', 'Actual', '% Used']].fillna(0)
joinDF.rename(columns={'Allocated Budget': 'Budget'}, inplace=True)
intCols=['Budget','Actual', '% Used', 'Budget1', 'Actual1', '% Used1']
joinDF.head()
```

# Previous Year Data

```python
data2=r'/kaggle//input/glbaq1/GLBA1.csv'
pdf=pd.read_csv(data2)
pDF=pdf[['GL Account', 'Allocated Budget', 'Actual', '% Used']].fillna(0)
pDF.rename(columns={'Allocated Budget': 'Budget1', 'Actual': 'Actual1','% Used': '% Used1'}, inplace=True)
pDF.head()
```

# Join Data with Previous Year Data

```python
joinDF=joinDF.merge(pDF, on='GL Account')
joinDF.head()
```

# Create New Tabs

```python
df1=joinDF[(joinDF['GL Account'] > 2_470_000) & (joinDF['GL Account']<2_480_000)]
df1.loc['Total Income']= df1[intCols].sum()
df1.head()
```

```python
df2=joinDF[(joinDF['GL Account'] > 2_450_000) & (joinDF['GL Account'] < 2_490_000)]
df2.loc['Total Income']= df2[intCols].sum()
df2.head()
```

```python
df3=joinDF[(joinDF['GL Account'] > 2_480_000) & (joinDF['GL Account'] < 2_490_000)]
df3.loc['Total Income']= joinDF[intCols].sum()
df3.head()
```
