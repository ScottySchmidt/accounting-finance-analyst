# Quarterly Report GLBA

# Current Quarter Data

```python
import pandas as pd
data=r'/kaggle/input/glbaqrt/GLBA.csv'
data=pd.read_csv(data)
data.head()
```

```python
from babel.numbers import format_currency

data=data[['GL Account', 'GL Description', 'Allocated Budget', 'Actual', '% Used']].fillna(0)
data.rename(columns={'Allocated Budget': 'Budget', 'GL Description': 'Revenue'}, inplace=True)
data.insert(5, " . ", " ") 

#CURRENCY:
#data["Budget"] = data["Budget"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
#data["Actual"] = data["Actual"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
#data['% Used'] = data['% Used'].apply(lambda x: "{0:.2f}%".format(x*100))
data.head()
```

# Previous Year Data

```python
#READ PREVIOUS QRT DATA:
data2=r'/kaggle//input/glbaq1/GLBA1.csv'
data2=pd.read_csv(data2)
data2=data2[['GL Account', 'Allocated Budget', 'Actual', '% Used']].fillna(0)
data2.head()
```

#### Rename Columns

```python
data2.rename(columns={'Allocated Budget': 'Budget1', 'Actual': 'Actual1','% Used': '% Used1'}, inplace=True)

#CHANGE CURRENCY:
#data2["Budget1"] = data2["Budget1"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
#data2["Actual1"] = data2["Actual1"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
#data2['% Used1'] = data2['% Used1'].apply(lambda x: "{0:.2f}%".format(x*100))
data2.head()
```

# Join Data with Previous Year Data

```python
revDF=data.merge(data2, on='GL Account')
revDF.head()
```

```python
revDF=revDF.rename(columns={'GL Account':'Object Code'}).reset_index(drop=True)
revDF.head()
```

# Split Revenue and Expense

```python
revDF['dept']  =   revDF['Object Code'].astype(str).str[:-3].astype(int)
revDF['Object Code']  = revDF['Object Code'].astype(str).str[-3:].astype(int)
revDF.head()
```

```python
revDF=revDF[revDF['Object Code']> 500]
revDF.head()
```

# Create New Tabs by Dept

```python
import warnings
warnings.filterwarnings("ignore")

intCols=['Budget','Actual', '% Used', 'Budget1', 'Actual1', '% Used1']
df1=revDF[(revDF['dept'] ==2471)]
df1.loc['Total Income']= df1[intCols].sum()
df1["Budget"] = df1["Budget"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
df1["Budget1"] = df1["Budget1"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
df1["Actual"] = df1["Actual"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
df1["Actual1"] = df1["Actual1"].apply(lambda x: format_currency(x, currency="USD", locale="en_US"))
#df1['% Used1'] = df1['% Used1'].apply(lambda x: "{0:.2f}%".format(x*100))
df1.head()
```

```python
df2=revDF[(revDF['dept']==2472)]
df2.loc['Total Income']= df2[intCols].sum()
df2.head()
```

```python
df3=revDF[(revDF['dept'] == 2473)]
df3.loc['Total Income']= revDF[intCols].sum()
df3.head()
```
