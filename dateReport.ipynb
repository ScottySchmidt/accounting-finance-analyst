# QuarterReport

```python
import numpy as np 
import pandas as pd 
import os

file=r'/kaggle/input/datareport/dataReport.csv'
df=pd.read_csv(file)
df.head()
```

```python
df=df[['GL Account', ' Date', ' YTD Debits', ' YTD Credits']]
df.head()
```

```python
df[[' YTD Debits', ' YTD Credits']]=df[[' YTD Debits', ' YTD Credits']].fillna(0)
df.head()
```

```python
df['DataTotal'] = df.apply((lambda x: x[' YTD Debits'] - x[' YTD Credits']), axis=1)
df.head()
```

```python
df=df.drop(columns=[' YTD Debits', ' YTD Credits'])
df.head()
```

```python
df=df.groupby(['GL Account', ' Date'])['DataTotal'].sum().reset_index()
df.head()
```

```python
df1=df[(df[' Date'] > '7/01/2021') & (df[' Date'] < '9/30/2021')].reset_index()
df1.head(40)
```

```python
df1=df1.groupby(['GL Account'])['DataTotal'].sum().reset_index()
df1.head(1000)
```
